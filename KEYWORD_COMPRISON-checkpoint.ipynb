```python
import pandas as pd
import numpy as np
import PyPDF2
import textract
import re
```

```python
filename ='C:/Users/USER/keyword_comparision/ALL5.pdf' 

pdfFileObj = open(filename,'rb')              
pdfReader = PyPDF2.PdfFileReader(pdfFileObj)  
num_pages = pdfReader.numPages                 
```

```python
num_pages
```

```python
count = 0
text = ""
                                                            
while count < num_pages:                       #The while loop will read each page
    pageObj = pdfReader.getPage(count)
    count +=1
    text += pageObj.extractText()
    
#Below if statement exists to check if the above library returned #words. 
#It's done because PyPDF2 cannot read scanned files.

if text != "":
    text = text
else:
    text = textract.process('C:/Users/USER/keyword_comparision/ALL5.pdf', method='tesseract', language='eng')
```

```python

text = text.encode('ascii','ignore').lower() #Lowercasing each word
```
